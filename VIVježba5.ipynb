```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from keras.datasets import boston_housing
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

def model_mreze(broj_atributa):
  model = Sequential(name="model")
  model.add(layers.Dense(64,activation='relu', input_shape=(broj_atributa,)))
  model.add(layers.Dense(64,activation='relu'))
  model.add(layers.Dense(1))
  return model

def smooth_curve(points, factor=0.9) :
  smoothed_points = []
  for point in points:
    if smoothed_points:
      previous = smoothed_points[-1]
      smoothed_points.append(previous * factor + point * (1-factor))
    else:
      smoothed_points.append(point)
  return smoothed_points

(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()

feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

df_train = pd.DataFrame(train_data, columns=feature_names)
df_train['TARGET'] = train_targets

df_test = pd.DataFrame(test_data, columns=feature_names)
df_test['TARGET'] = test_targets

scaler = MinMaxScaler()
scaler.fit(train_data)

scaled_train_data = scaler.transform(train_data)
scaled_test_data = scaler.transform(train_data)

num_features = scaled_train_data.shape[1]

model = model_mreze(num_features)

model.summary()

model.compile(optimizer='adam', loss ='mse', metrics=['mae'])

num_epochs = 100
batch_size = 1
validation_split = 0.1

history = model.fit(scaled_train_data,
                        train_targets,
                        epochs=num_epochs,
                        batch_size=batch_size,
                        validation_split=validation_split,
                        verbose=1)

loss = history.history['loss']
val_loss = history.history['val_loss']
mae = history.history['mae']
val_mae = history.history['val_mae']

epochs_range = range(1, len(loss) + 1)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, 'bo-', label='Trening Loss (MSE)', markersize=3)
plt.plot(epochs_range, val_loss, 'ro-', label='Validacijski Loss (MSE)', markersize=3)
plt.title('Funkcija gubitka (MSE) tijekom treniranja')
plt.xlabel('Epohe')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_range, mae, 'bo-', label='Trening MAE', markersize=3)
plt.plot(epochs_range, val_mae, 'ro-', label='Validacijski MAE', markersize=3)
plt.title('Metrika (MAE) tijekom treniranja')
plt.xlabel('Epohe')
plt.ylabel('MAE (u tisućama $)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

smooth_val_mae = smooth_curve(val_mae[10:])
epochs_for_smoothed = range(11, len(val_mae) + 1)
plt.figure(figsize=(10, 6))
plt.plot(epochs_for_smoothed,
             smooth_val_mae,
             'go-',
             label=f'Izglađeni Validacijski MAE (od epohe {10 + 1})',
             markersize=4)


plt.title('Izglađena Krivulja Validacijske MAE Metrike')
plt.xlabel('Epohe')
plt.ylabel('Izglađeni MAE (u tisućama $)')
plt.legend()
plt.grid(True)
plt.ylim(bottom=0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from keras.datasets import boston_housing
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

def model_mreze(broj_atributa):
  model = Sequential(name="model")
  model.add(layers.Dense(64,activation='relu', input_shape=(broj_atributa,)))
  model.add(layers.Dense(64,activation='relu'))
  model.add(layers.Dense(1))
  return model

def smooth_curve(points, factor=0.9) :
  smoothed_points = []
  for point in points:
    if smoothed_points:
      previous = smoothed_points[-1]
      smoothed_points.append(previous * factor + point * (1-factor))
    else:
      smoothed_points.append(point)
  return smoothed_points


(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()


scaler = MinMaxScaler()
scaler.fit(train_data)


scaled_train_data = scaler.transform(train_data)
scaled_test_data = scaler.transform(test_data)


num_features = scaled_train_data.shape[1]
model = model_mreze(num_features)
print("--- Model Summary ---")
model.summary()


model.compile(optimizer='adam', loss ='mse', metrics=['mae'])


num_epochs = 500
batch_size = 1
validation_split = 0.1

print("\n--- Training Model ---")
history = model.fit(scaled_train_data,
                        train_targets,
                        epochs=num_epochs,
                        batch_size=batch_size,
                        validation_split=validation_split,
                        verbose=1)


loss = history.history['loss']
val_loss = history.history['val_loss']
mae = history.history['mae']
val_mae = history.history['val_mae']

epochs_range = range(1, len(loss) + 1)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, 'bo-', label='Trening Loss (MSE)', markersize=3)
plt.plot(epochs_range, val_loss, 'ro-', label='Validacijski Loss (MSE)', markersize=3)
plt.title('Funkcija gubitka (MSE) tijekom treniranja')
plt.xlabel('Epohe')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_range, mae, 'bo-', label='Trening MAE', markersize=3)
plt.plot(epochs_range, val_mae, 'ro-', label='Validacijski MAE', markersize=3)
plt.title('Metrika (MAE) tijekom treniranja')
plt.xlabel('Epohe')
plt.ylabel('MAE (u tisućama $)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.suptitle('Rezultati treniranja', y=1.02)
plt.show()


points_to_skip = 10
if len(val_mae) > points_to_skip:
    smooth_val_mae = smooth_curve(val_mae[points_to_skip:])
    epochs_for_smoothed = range(points_to_skip + 1, len(val_mae) + 1)

    plt.figure(figsize=(10, 6))
    plt.plot(epochs_for_smoothed, smooth_val_mae, 'go-',
             label=f'Izglađeni Validacijski MAE (od epohe {points_to_skip + 1})',
             markersize=4)
    plt.title('Izglađena Krivulja Validacijske MAE Metrike')
    plt.xlabel('Epohe')
    plt.ylabel('Izglađeni MAE (u tisućama $)')
    plt.legend()
    plt.grid(True)
    plt.ylim(bottom=0)
    plt.show()
else:
     print(f"\nNema dovoljno epoha ({len(val_mae)}) za prikaz izglađene krivulje (potrebno > {points_to_skip}).")




print("\n--- Evaluacija modela na neviđenom testnom skupu ---")

test_loss, test_mae = model.evaluate(scaled_test_data, test_targets, verbose=0)

print(f"Rezultati evaluacije na testnom skupu:")
print(f"  - Funkcija gubitka (MSE): {test_loss:.4f}")
print(f"  - Metrika (MAE):        {test_mae:.4f}")


mae_in_dollars = test_mae * 1000
print(f"\nNa osnovu MAE metrike, model na testnom skupu u prosjeku")
print(f"odstupa od stvarne cijene kuće za: ${mae_in_dollars:.2f}")
```
